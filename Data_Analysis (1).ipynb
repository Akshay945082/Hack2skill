```python
# Plot a pie chart of the overall event ratings
rating_counts = matched_feedbacks['Rate your experience with the overall event'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Ratings for Overall Event")
plt.show()
```

```python
# Count occurrences for each city
city_counts = matched_feedbacks['City'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6,6))
plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Feedbacks by City")
plt.show()
```

```python

activity_columns = ['Rate the activity-1', 'Rate the following activity-2', 'Rate the following activity-3', 'Rate the following activity-4']
activity_ratings =matched_feedbacks[activity_columns].apply(pd.to_numeric, errors='coerce')
activity_ratings_mean = activity_ratings.mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=activity_ratings_mean.index, y=activity_ratings_mean.values)
plt.title("Average Ratings for Activities")
plt.ylabel('Average Rating')
plt.xlabel('Activities')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gender distribution
gender_distribution = matched_feedbacks["Gender"].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 6))
gender_distribution.plot.pie(autopct="%1.1f%%", startangle=140, colors=["red", "green"])
plt.title("Gender Distribution")
plt.show()
```

```python
# Overall experience ratings
experience_ratings = matched_feedbacks["Rate your experience with the overall event"].value_counts()

# Plot bar chart
plt.figure(figsize=(8, 6))
experience_ratings.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Overall Experience Ratings")
plt.xlabel("Ratings")
plt.ylabel("Number of Responses")
plt.xticks(rotation=0)
plt.show()
```
